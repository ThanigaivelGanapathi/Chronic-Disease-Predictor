```python
#Loading Libraries
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
dataset = pd.read_csv("CKD.csv")
dataset
```

```python
#Data Preprocessing

dataset = pd.get_dummies(dataset,dtype = int, drop_first  = True)
dataset
```

```python
dataset.columns
```

```python
#splitting Input and Output
independent = dataset.drop("classification_yes",axis = 1)
dependent = dataset[["classification_yes"]]
```

```python
#Splitting Training and Testing Dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size = 0.3,random_state = 0)
```

Decision Tree Model

```python
#Decision Tree Model

dt_params = {
    "criterion"    : ["gini", "entropy", "log_loss"],
    "splitter"     : ["best", "random"],
    "max_features" : ["sqrt", "log2"]
}

dt_classifier = GridSearchCV(DecisionTreeClassifier(),dt_params, refit = True)  
dt_classifier.fit(x_train,y_train) 
print(dt_classifier.best_params_)
```

```python
#Predicting
y_dt_pred = dt_classifier.predict(x_test)

#finding Confustion Matrix
dt_confusion_mtx = confusion_matrix(y_test,y_dt_pred)
print(dt_confusion_mtx)

#finding Classification Report
dt_classification_rpt = classification_report(y_test,y_dt_pred)
print(dt_classification_rpt)
```

Random Forest Model

```python
#Random Forest Model

rf_params = {
    "criterion"    : ["gini", "entropy", "log_loss"],
    "n_estimators" : [50,100,250,500,1000],
    "max_features" : ["sqrt", "log2",None]
}

rf_classifier = GridSearchCV(RandomForestClassifier(),rf_params, refit = True) 
print(rf_classifier)
rf_classifier.fit(x_train,y_train)
print(rf_classifier.cv_results_)
print(rf_classifier.best_params_)
```

```python
#Predicting
y_rf_pred = rf_classifier.predict(x_test)

#finding Confustion Matrix
rf_confusion_mtx = confusion_matrix(y_test,y_rf_pred)
print(rf_confusion_mtx)

#finding Classification Report
rf_classification_rpt = classification_report(y_test,y_rf_pred)
print(rf_classification_rpt)
```

Logistic Regression Model

```python
#Logistic Regression Model

lg_params = {       
    "C"   : [1.0,5.0,10.0],   
}

lg_classifier = GridSearchCV(LogisticRegression(),lg_params, refit = True)  
lg_classifier.fit(x_train,y_train) 
print(lg_classifier.best_params_)
```

```python
#Predicting
y_lg_pred = lg_classifier.predict(x_test)

#finding Confustion Matrix
lg_confusion_mtx = confusion_matrix(y_test,y_lg_pred)
print(lg_confusion_mtx)

#finding Classification Report
lg_classification_rpt = classification_report(y_test,y_lg_pred)
print(lg_classification_rpt)
```

KNN Model

```python
#Knn Model

knn_params = {    
    "n_neighbors" : [5,10,100,150,200],
    "algorithm"   : ["auto", "ball_tree", "kd_tree", "brute"],
    "metric" : ["minkowski"]    
}

knn_classifier = GridSearchCV(KNeighborsRegressor(),knn_params, refit = True)  
knn_classifier.fit(x_train,y_train) 
print(knn_classifier.best_params_)
```

```python
#Predicting
y_knn_pred = knn_classifier.predict(x_test)
 
threshold = 0.5
y_knn_pred_binary = (y_knn_pred > threshold).astype(int)
    
#finding Confustion Matrix
knn_confusion_mtx = confusion_matrix(y_test,y_knn_pred_binary)
print(knn_confusion_mtx)

#finding Classification Report
knn_classification_rpt = classification_report(y_test,y_knn_pred_binary)
print(knn_classification_rpt)
```
